```python
import s3fs
import xarray as xr
import fsspec
```

## Load Data

### OISST

```python
endpoint_url = 'https://ncsa.osn.xsede.org'
fs_osn = s3fs.S3FileSystem(anon=True, client_kwargs={'endpoint_url': endpoint_url},) 

path = "Pangeo/pangeo-forge/noaa_oisst/v2.1-avhrr.zarr"
ds = xr.open_zarr(fs_osn.get_mapper(path), consolidated=True)
oisst = ds.sst.isel(zlev=0).drop('zlev')
```

### MUR

```python
file_location = 's3://mur-sst/zarr'

ikey = fsspec.get_mapper(file_location, anon=True)

mur_full = xr.open_zarr(ikey,consolidated=True)
mur = mur_full['analysed_sst']
```

## Subset

```python
min_lat = 41.375
max_lat = 42.625
min_lon = -74.375
max_lon = -69.625

# OISST is organized 0->360 instead of -180 -> 180
oisst_min_lon = min_lon + 360
oisst_max_lon = max_lon +360
```

```python
# Subset OISST
nw_oisst = oisst.sel(
    time='2002-06-01T12', 
    lat=slice(min_lat, max_lat), 
    lon=slice(oisst_min_lon, oisst_max_lon)
)
```

```python
# Subset MUR
nw_mur = mur.sel(
    time='2002-06-01', 
    lat=slice(min_lat, max_lat), 
    lon=slice(min_lon, max_lon)
  )
```

## Plot

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())

p = nw_oisst.plot(ax=ax)

ax.coastlines()
```

```python
ax = plt.axes(projection=ccrs.PlateCarree())

p = nw_mur.plot(ax=ax)

ax.coastlines()
```
